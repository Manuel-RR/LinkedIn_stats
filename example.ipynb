**Visualize Progress:**

```python
# Plot cumm progress:
import matplotlib.pyplot as plt

# create subplots
fig, axes = plt.subplots(nrows=2, ncols=1)

# plot the first graph
engagement['Impressions'].cumsum().plot(ax=axes[0])
axes[0].set_ylabel('Impressions')  # add label to y-axis

# plot the second graph
followers['New followers'].cumsum().plot(ax=axes[1])
axes[1].set_ylabel('New followers')  # add label to y-axis

# set title for the entire figure
fig.suptitle('Cumulative Sum Graphs')

# display the plot
plt.show()
```
